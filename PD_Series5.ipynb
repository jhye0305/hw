```python
#4.7.1
import pandas as pd
import numpy as np

np.random.seed(0)
df2 = pd.DataFrame({
    'key1': ['A', 'A', 'B', 'B', 'A'],
    'key2': ['one', 'two', 'one', 'two', 'one'],
    'data1': [1, 2, 3, 4, 5],
    'data2': [10, 20, 30, 40, 50]
})
df2.groupby(df2.key1)["data1"].sum()
```

```python
#4.7.2
import seaborn as sns
iris = sns.load_dataset("iris")
def peak_to_peak_ratio(x):
    return x.max() / x.min()

iris.groupby(iris.species).agg(peak_to_peak_ratio)
iris.groupby(iris.species).describe().T
```

```python
import seaborn as sns
iris = sns.load_dataset("iris")

mean_by_species = iris.groupby('species').mean()

print(mean_by_species[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])
```

```python
import seaborn as sns
import pandas as pd
titanic = sns.load_dataset("titanic")

age_groups = pd.qcut(titanic['age'], q=3, labels=['Q1', 'Q2', 'Q3'])

titanic['survived'] = titanic['survived'].astype(bool)
survival_rate = titanic.groupby(['sex', 'class', age_groups])['survived'].mean().unstack()
print(survival_rate)
```
