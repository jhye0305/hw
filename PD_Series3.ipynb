```python
#4.4.3
import seaborn as sns
import pandas as pd
titanic = sns.load_dataset("titanic")
average_age_all_passengers = titanic['age'].mean()
print("Average Age of All Passengers:", average_age_all_passengers)
average_age_female_passengers = titanic[titanic['sex'] == 'female']['age'].mean()
print("Average Age of Female Passengers:", average_age_female_passengers)
average_age_female_1st_class_passengers = titanic[(titanic['sex'] == 'female') & (titanic['class'] == 'First')]['age'].mean()
print("Average Age of Female Passengers in 1st Class:", average_age_female_1st_class_passengers)
```

```python
#4.4.4
titanic["adult/child"] = titanic.apply(lambda r: "adult" if r.age >= 20 else "child", axis=1)
def category1(row):
    if row['age'] >= 20:
        return row['sex']
    else:
        return 'child'
titanic['category1'] = titanic.apply(category1, axis=1)
titanic.tail()
```

```python
#4.4.5 ???
import seaborn as sns
import pandas as pd
titanic = sns.load_dataset("titanic")

mean_age = titanic['age'].mean()
titanic['age'] = titanic['age'].fillna(mean_age)

print(titanic.head())
```

```python
#4.4.6
titanic['age'] = titanic['age'].astype(int)
titanic['category2'] = titanic.sex+titanic.age.astype(str)
titanic.tail()
```

```python
#4.4.7
bins = [1, 20, 30, 50, 70, 100]
labels = ["미성년자", "청년", "중년", "장년", "노년"]
titanic["연령대"] = pd.cut(titanic.age, bins, labels=labels)
titanic_age = pd.DataFrame(titanic.연령대.value_counts(), columns=["연령대"])
titanic_age["비율"] = titanic_age.apply(lambda i : i.연령대 / titanic_age.연령대.sum(), axis=1)
titanic.tail()
```

```python
#4.4.8
titanic["category3"] = titanic.apply(lambda i : i.연령대 if i.age < 20 else i.연령대+i.sex, axis=1)
titanic.tail()
```
